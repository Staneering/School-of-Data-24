```python
import pandas as pd

# Assuming your DataFrame is called 'numeric_columns'
def identify_problematic_columns(df):
    """Identifies columns causing ValueError during numeric conversion.

    Args:
        df: The DataFrame to analyze.

    Returns:
        A list of column names causing the ValueError.
    """
    problematic_columns = []
    for column in df.columns:
        try:
            pd.to_numeric(df[column].str.split('[+-]').str[0])  # Mimicking the problematic operation
        except ValueError:
            problematic_columns.append(column)
    return problematic_columns

# Example usage:
problematic_cols = identify_problematic_columns(numeric_df)
print("Problematic columns:", problematic_cols)
```
