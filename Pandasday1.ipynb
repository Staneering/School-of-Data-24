```python
import numpy as np
import random as rd
```

```python
#create a dataframe
#student id,name,maths,english,physics,chemistry
#create a subject list
#create a name list

student_dictionary = {'maths':[],'english':[],'physics':[],'chemistry':[]}
list_length = 100
name_list = []
subject_list = ['maths','english','physics','chemistry']
for i in range(1,101):
    name_list.append('student'+str(i))
print(name_list)


```

```python
#create random values from 0 to 100 for each subject

for subject in student_dictionary:
    student_dictionary[subject] = np.random.randint(0,101,list_length)
print(student_dictionary)
```

```python
student_dictionary['name'] = name_list
print(student_dictionary)
```

```python
import pandas as pd
df = pd.DataFrame(student_dictionary)
print(df)
```

```python
column_order = ['name', 'maths', 'english', 'physics', 'chemistry']


df = df[column_order]
print(df)
```
