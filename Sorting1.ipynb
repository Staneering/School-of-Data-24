## Introduction

When a dataset only has a few data points, it's very easy to manually apply operations to the entire dataset. However, good datasets are very rarely this small. The Pandas library in Python allows us to use vectorised operations (i.e. operations across multiple data points simultaneously), making it much easier to work with large amounts of data.

In this train, we look at how to manipulate data within dataframes. Let's start by importing the Pandas library and the dataset.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

```

```python
df = pd.read_csv('fifa (1).csv')
```

```python
df.head()
```

```python
df.columns
```

Sorting

One basic operation to perform when working with a large dataset is to sort the values using some criteria. For dataframes, sorting can be achieved by using the sort_values() function. To sort data entries by a specific column(s), we need to pass this column(s) to the function. Let's look at some examples:

```python
df.sort_values('Overall',ascending=False)
```

```python
df.isna().sum()
```

```python
df['Age'].unique()
```

```python
df['Age'].replace(32,np.nan,inplace=True)
```

```python
df['Age']
```

```python
df.sort_values(by = 'Age',ascending = False)
```

```python
df['Age'].fillna(32,inplace=True)
```

Filtering

```python
df[df['Age']==32]
```

```python
df['Rating Per Age']=df['Overall']/df['Age']

df['Rating Per Age']
```

```python
df['Rating per Age2']=df.apply(lambda x: x['Overall']/x['Age'],axis = 1)
```

```python
df['Rating per Age2']
```

# Deleting Columns

```python
df.drop(['Rating per Age2','Rating Per Age'],axis=1,inplace=True)
```

```python
df.columns
```

```python
df.head()
```

# Grouping Values

```python
df[df['Age']==16]
```

```python
df.groupby('Age')['Overall'].mean()
```

```python
max_age_group =df.groupby('Age')['Overall'].max()
```

```python
max_age_group.plot(kind = 'bar',figsize=(10,6))

plt.title('Bar Chart of Maximum Overall Rating by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Maximum Overall Rating')
plt.show()
```
